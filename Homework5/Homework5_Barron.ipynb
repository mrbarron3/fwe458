```python
# imports
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# define x and y variables for use in problem 1
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
# define the figure
fig, ax = plt.subplots()

# plot the data
ax.plot(x, y, linestyle = "dashed", color= "blue")

# add in labels
ax.set_title("Sin Graph")
ax.set_xlabel("X")
ax.set_ylabel("sin(X)")

# supress the string output
plt.show()
```

```python
# load the tips dataset for all future questions
tips = sns.load_dataset("tips")

tips.head()
```

```python
# create the histogram and overlay a kde
sns.histplot(data = tips,
            x = "total_bill",
             bins = 20,
             color = "green",
             kde = True)

# add in labels
plt.xlabel("Total Bill Amount")
plt.title("Frequency of Restaurant Bill Amounts")

# supress the string output
plt.show()
```

```python
# prepare the placement of figures to be plotted
fig, axs = plt.subplots(1, 2, figsize = (10, 5))

# make the box plot in the first slot
sns.boxplot(data = tips, x = "time", y = "tip", ax = axs[0], color = "paleturquoise")
axs[0].set_title("Tip Distribution (Box)")
axs[0].set_xlabel("Time of Day")
axs[0].set_ylabel("Tip Amount")

# make the violin plot in the second slot
sns.violinplot(data = tips, x = "time", y = "tip", ax = axs[1], color = "lightcoral")
axs[1].set_title("Tip Distribution (Violin)")
axs[1].set_xlabel("Time of Day")
axs[1].set_ylabel("Tip Amount")

# supress the string output
plt.show()
```

```python
# create the scatter plot with a regression line
sns.lmplot(tips, x = "total_bill", y = "tip", hue = "smoker")

# add in labels
plt.title("Relationship between Restaurant Bill Amount and Tip Amount")
plt.xlabel("Bill Amount")
plt.ylabel("Tip Amount")

# supress the string output
plt.show()
```

```python
# calculate the correlation matrix using only numeric columns
corr_matrix = tips.corr(numeric_only = True)

# plot the heatmap with annotations
# set the min and max values for the colormap to be the min and absolute max correlation values possible
sns.heatmap(corr_matrix, cmap = 'BrBG', vmin = 0, vmax = 1, annot = True)

# supress the string output
plt.show()
```

```python
# create the pairplot
g = sns.pairplot(data = tips,
             hue = "time")

# add in KDE plots along the diagonals
g.map_diag(sns.kdeplot)

# supress the string output
plt.show()
```
