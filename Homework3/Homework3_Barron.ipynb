```python
# imports
import pandas as pd
from google.colab import drive
```

```python
# Import Data
# mounting drive
drive.mount('/content/drive')

# specify the path to the file
filedir = "/content/drive/MyDrive/Spring '25/FWE458/"

fname = filedir + "MaunaLoaCO2_monthly_sitename.csv"

# read in the csv file
df = pd.read_csv(fname)

df.head()
```

```python
# Filter Data
# only keep rows NOT from the years 1958 and 2022
df_years = df[(df["Year"] != 1958) & (df["Year"] != 2022)]
print(df_years.head())

# keep only rows with non-negative values in numofdays and stdofdays columns
df_pos = df_years[(df_years["numofdays"] >= 0) & (df_years["stdofdays"] >= 0)]
print(df_pos.head())
```

```python
# Annual Statistics
# find the mean of the monthly CO2 concentrations for each year
annual_average = df_pos.groupby("Year")["Monthly_average"].mean()
annual_average.head()
```

```python
# find the mean of the monthly deseasonalized CO2 concentrations for each year
deseasonalized_annual_average = df_pos.groupby("Year")["deseasonalized"].mean()
deseasonalized_annual_average.head()
```

```python
# find the standard deviation of the monthly CO2 concentrations for each year
std_of_months = df_pos.groupby("Year")["Monthly_average"].std()
std_of_months.head()
```

```python
# Custom Function
# define the function with a single parameter (co_level)
def co2_categories(co_level):
  # check if the co_level is between certain thresholds and categorize it
  if co_level < 350:
    return "Low"
  elif 350 <= co_level <= 400:
    return "Moderate"
  else:
    return "High"
```

```python
# apply the previous function on the monthly CO2 concentration column to create a new column with categories
df_pos["CO2_category"] = df_pos["Monthly_average"].apply(co2_categories)
df_pos.head()
```

```python
fname = filedir + "Barron_MaunaLoaCO2_annual_Hw3.csv"
# export the dataframe to my google drive
df_pos.to_csv(fname)
```
